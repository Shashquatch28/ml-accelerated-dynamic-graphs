# Dijkstra's Algorithm with NetworkX

This notebook demonstrates how to use the networkx library to create weighted graphs and compute shortest paths using Dijkstra's algorithm. It serves as a reference implementation to validate and understand the algorithm before writing a custom version.

The notebook covers:

- Graph creation and visualization

- Computing shortest paths from a single source node

- Displaying path lengths and node sequences

# Import Libraries

```python
import networkx as nx
import matplotlib.pyplot as plt
```

# Test Graph

Here, we create an undirected graph G and add nodes connected by weighted edges. The weights represent the cost or distance between nodes and are used by Dijkstra's algorithm to find shortest paths.

```python
G = nx.Graph()

# Add weighted edges
G.add_edge('A', 'B', weight=4)
G.add_edge('A', 'C', weight=2)
G.add_edge('B', 'C', weight=1)
G.add_edge('B', 'D', weight=5)
G.add_edge('C', 'D', weight=8)
G.add_edge('C', 'E', weight=10)
G.add_edge('D', 'E', weight=2)
G.add_edge('D', 'Z', weight=6)
G.add_edge('E', 'Z', weight=3)
```

## Visualization

This cell visualizes the graph structure using NetworkXâ€™s drawing functions and Matplotlib. Nodes and edges are displayed along with edge weights to help intuitively understand the graph before running shortest path computations.

```python
pos = nx.spring_layout(G) 

# Nodes and edges
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1000, font_size=14)

# Edge labels (weights)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title('Test Graph Visualisation')
plt.show()
```

## Compute Shortest Path

We use NetworkXâ€™s single_source_dijkstra function to compute shortest path lengths and routes from a specified source node to all reachable nodes. The results include both minimum distances and node sequences for these paths, which are then printed for verification.

```python
source_node = 'A'
lengths, paths = nx.single_source_dijkstra(G, source=source_node, weight='weight')

print(f"Shortest path lengths from node '{source_node}':")
for target, dist in lengths.items():
    print(f"{source_node} -> {target}: {dist}")

print(f"\nShortest paths from node '{source_node}':")
for target, path in paths.items():
    print(f"{source_node} -> {target}: {path}")
```
